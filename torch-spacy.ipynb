```python
import spacy
import torch

print(f"spaCy version: {spacy.__version__}")
print(f"PyTorch version: {torch.__version__}")
print(f"CUDA available: {torch.cuda.is_available()}")
print(f"CUDA device count: {torch.cuda.device_count()}")
if torch.cuda.is_available():
    print(f"CUDA device name: {torch.cuda.get_device_name(0)}")
    print(f"CUDA version in use:{torch.version.cuda}")

# Correct way to try to use the GPU with spaCy v3+
try:
    spacy.prefer_gpu()  # Or spacy.require_gpu()
    nlp = spacy.load("en_core_web_lg")  # Load the model *after* prefer_gpu
    #print(f"spaCy using GPU: {nlp.device_type == 'cuda'}")  # Check device type -- Removed
    #print(f"spaCy device: {nlp.device}") -- Removed
    if spacy.util.has_gpu(): # correct check
        print("spaCy is using GPU")
    else:
        print("spaCy is using CPU")
except Exception as e:
    print(f"Error loading spaCy with GPU: {e}")

x = torch.rand(5, 3)
print(x)
```
